### Multiclass Logistic Regression

Multiclass Logistic Regression, also known as **Softmax Regression**, is an extension of logistic regression that is used for classification problems where the target variable has more than two classes. It generalizes logistic regression to handle multiple classes instead of just a binary outcome.

In multiclass logistic regression, the goal is to model the probability that a given instance belongs to each of the possible classes. The model uses the **softmax function** to output the probability distribution over multiple classes.

#### The Softmax Function

The softmax function is used to convert the raw output of the model into probabilities that sum to 1. Given a set of raw scores (also known as logits), the softmax function is defined as:

$$
P(y = k \mid \mathbf{x}) = \frac{e^{z_k}}{\sum_{j=1}^{K} e^{z_j}}
$$

Where:
- \( P(y = k \mid \mathbf{x}) \) is the probability that the instance \( \mathbf{x} \) belongs to class \( k \),
- \( z_k \) is the raw score (logit) for class \( k \),
- \( K \) is the number of classes,
- The denominator is the sum of the exponentiated logits for all classes.

#### Model Representation

In multiclass logistic regression, we model the probability of each class \( k \) as a function of the input features \( \mathbf{x} \). The raw scores \( z_k \) for each class are given by a linear combination of the input features:

$$
z_k = \mathbf{w}_k^T \mathbf{x} + b_k
$$

Where:
- \( \mathbf{w}_k \) is the weight vector for class \( k \),
- \( \mathbf{x} \) is the input feature vector,
- \( b_k \) is the bias term for class \( k \).

Thus, the probability for each class is:

$$
P(y = k \mid \mathbf{x}) = \frac{e^{\mathbf{w}_k^T \mathbf{x} + b_k}}{\sum_{j=1}^{K} e^{\mathbf{w}_j^T \mathbf{x} + b_j}}
$$

#### Cost Function (Cross-Entropy)

To train the model, we use a **multiclass cross-entropy loss** function, which is a generalization of the binary cross-entropy loss used in binary logistic regression. The cost function is:

$$
J(\mathbf{W}, \mathbf{b}) = - \frac{1}{m} \sum_{i=1}^{m} \sum_{k=1}^{K} y^{(i)}_k \log P(y^{(i)}_k \mid \mathbf{x}^{(i)})
$$

Where:
- \( m \) is the number of training examples,
- \( y^{(i)}_k \) is the indicator (0 or 1) of whether the true class of example \( i \) is class \( k \),
- \( P(y^{(i)}_k \mid \mathbf{x}^{(i)}) \) is the predicted probability for class \( k \) for example \( i \).

#### Gradient Descent

To minimize the cost function, we use **gradient descent** to iteratively update the weights and biases for each class. The gradient for each weight \( \mathbf{w}_k \) and bias \( b_k \) is computed using the chain rule:

For each weight \( \mathbf{w}_k \), the gradient is:

$$
\frac{\partial J}{\partial \mathbf{w}_k} = \frac{1}{m} \sum_{i=1}^{m} \left( P(y^{(i)}_k \mid \mathbf{x}^{(i)}) - y^{(i)}_k \right) \mathbf{x}^{(i)}
$$

For each bias \( b_k \), the gradient is:

$$
\frac{\partial J}{\partial b_k} = \frac{1}{m} \sum_{i=1}^{m} \left( P(y^{(i)}_k \mid \mathbf{x}^{(i)}) - y^{(i)}_k \right)
$$

Where \( P(y^{(i)}_k \mid \mathbf{x}^{(i)}) \) is the predicted probability for class \( k \) for example \( i \).

#### Output

The output of multiclass logistic regression is a set of learned weights \( \mathbf{W} = [\mathbf{w}_1, \mathbf{w}_2, \dots, \mathbf{w}_K] \) and biases \( \mathbf{b} = [b_1, b_2, \dots, b_K] \), which can be used to predict the class probabilities for new data.

To classify a new instance, we compute the probability for each class and choose the class with the highest probability:

$$
\hat{y} = \arg\max_k P(y = k \mid \mathbf{x})
$$

### Key Points:
- **Multiclass Logistic Regression** is an extension of logistic regression for classification problems with more than two classes.
- It uses the **softmax function** to compute the probability distribution over multiple classes.
- The **multiclass cross-entropy loss** is used to train the model.
- The model is optimized using **gradient descent**.
- The output is the class with the highest probability for a given input.

### Mathematics Behind One-vs-Rest (OvR) in Logistic Regression

For **One-vs-Rest (OvR)**, logistic regression models a binary classification problem for each class. If there are \( K \) classes, \( K \) separate logistic regression classifiers are trained. Let’s break this down step by step:

---

### **1. Binary Logistic Regression Model**
For each class \( k \), the logistic regression model estimates the probability that an input \( \mathbf{x} \) belongs to class \( k \) as follows:
$$
P(y = k \mid \mathbf{x}) = \sigma(\mathbf{w}_k^\top \mathbf{x} + b_k)
$$

where:
- $$ (\sigma(z) )$$ is the sigmoid function, defined as:
  $$
  \sigma(z) = \frac{1}{1 + e^{-z}}
  $$
- $$ \mathbf{w}_k $$ is the weight vector for class $$  k $$,
- $$ b_k  $$ is the bias term for class $$ k $$,
-$$ \mathbf{x} $$ is the input feature vector.

The decision boundary for class $$ k $$ is determined by solving:
$$
\mathbf{w}_k^\top \mathbf{x} + b_k = 0
$$

---

### **2. Training OvR Classifiers**
For a dataset with \( N \) samples \( \{ (\mathbf{x}_i, y_i) \}_{i=1}^N \):
- For each class \( k \), create a binary classification problem:
  - Label samples belonging to class \( k \) as positive (\( y = 1 \)).
  - Label all other samples (\( y \neq k \)) as negative (\( y = 0 \)).



# Mathematics Behind One-vs-Rest (OvR)

### 1. Binary Logistic Regression Model
For each class \( k \), the probability of belonging to class \( k \) is given by:
$$
P(y = k \mid \mathbf{x}) = \sigma(\mathbf{w}_k^\top \mathbf{x} + b_k)
$$
where:
$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

### 2. Training OvR Classifiers
For a dataset with \( N \) samples \( \{ (\mathbf{x}_i, y_i) \}_{i=1}^N \), the log-loss for each class \( k \) is:
$$
\mathcal{L}_k = -\frac{1}{N} \sum_{i=1}^N \left[ y_{i,k} \log(P(y = k \mid \mathbf{x}_i)) + (1 - y_{i,k}) \log(1 - P(y = k \mid \mathbf{x}_i)) \right]
$$

### 3. Making Predictions
After training \( K \) classifiers, for a new input \( \mathbf{x} \), compute probabilities for each class:
$$
P_k = P(y = k \mid \mathbf{x}) = \sigma(\mathbf{w}_k^\top \mathbf{x} + b_k), \quad \forall k \in \{1, 2, \ldots, K\}
$$
The predicted class is:
$$
\hat{y} = \arg\max_k P_k
$$


```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
encoder = LabelEncoder()
df['species'] = encoder.fit_transform(df['species'])
```

```python
df.head()
```

```python
df = df[['sepal_length','petal_length','species']]
```

```python
df.head()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
clfo = LogisticRegression(multi_class='ovr')
clfm = LogisticRegression(multi_class='multinomial')
```

```python
clfo.fit(X_train,y_train)
clfm.fit(X_train,y_train)
```

```python
yo_pred = clfo.predict(X_test)
ym_pred = clfm.predict(X_test)
```

```python
print(accuracy_score(y_test,yo_pred))
print(accuracy_score(y_test,ym_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test,yo_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test,ym_pred))
```

```python
query = np.array([[3.4,2.7]])
clfo.predict_proba(query)
```

Use One-vs-Rest (OVR) when:
    

1. Classes are Non-Mutually Exclusive: OVR is appropriate if an instance can belong to more than one class, as each classifier provides an independent probability for each class.

2. Dealing with Imbalanced Data: OVR might perform better when class distribution is highly imbalanced since each class gets a dedicated model.



Use Multinomial Logistic Regression (SoftMax Regression) when:



1. Computational Efficiency is Required: Softmax Regression is generally more efficient for large datasets and a high number of classes.

2. Classes are Mutually Exclusive: SoftMax Regression is a good choice when each instancecan only belong to one class. The SoftMax function provides a set of probabilities that sum to 1, fitting well with mutually exclusive classes.

3. Interpretability is Important: The probabilities output by SoftMax Regression are more interpretable than those from OVR, as they always sum to 1. This can make model predictions easier to explain.
